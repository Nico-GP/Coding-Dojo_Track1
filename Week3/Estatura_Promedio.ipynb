Se agrupan las alturas de cada atleta y su año de participacion en las olimpiadas

```python
df.groupby(["ID", "Year"])[["Height"]].mean()
```

Altura promedio de todos los atletas cada año de las olimpiadas

```python
H_mean = df.groupby(["Year"])[["Height"]].mean()
H_mean
```

```python
import matplotlib.pyplot as plt
```

Grafico de alturas promedio por año

```python
plt.style.use("ggplot")
plt.plot(H_mean);
```
