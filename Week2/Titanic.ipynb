2. Porcentaje de supervivencia

```python
Sup = titanic["Survived"].mean()
Supervivencia = Sup * 100
print(Supervivencia)
```

Sobrevivio el 38,4% de los pasajeros

3. Genero de los supervivientes

```python
titanic.groupby(["Sex"])[["Survived"]].sum()
```

Sobrevivieron 233 mujeres y 109 hombres

4. Porcentaje de supervivencia con pasaje menor a 10

```python
Sobrevivientes = titanic["Survived"] == 1
Menos_10 = titanic["Fare"] < 10

Sob_Men_10 = len(titanic[Sobrevivientes & Menos_10]) / len(titanic["Survived"])
Porc_Sob_Men_10 = Sob_Men_10 * 100
Porc_Sob_Men_10
```

7,5% de los sobrevivientes tenian un pasaje menor a 10

5. y 6. Edad promedio de los **NO** sobrevivientes y sobrevivientes

```python
titanic.groupby(["Survived"])[["Age"]].mean()
```

La edad promedio es de:
* 30,6 años para los **no** sobrevivientes
* 28,3 años para los sobrevivientes 

7. Edad promedio de sobrevivientes y no sobrevivientes agrupados por genero

```python
titanic.groupby(["Sex","Survived"])[["Age"]].mean()
```

La edad promedio de:
* Sobrevivientes:
  * Mujeres 28,8 años
  * Hombres 27,3 años
* No Sobrevivientes:
  * Mujeres 25,0 años
  * Hombres 31,6 años
