Se importa pandas 

```python
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/CodingDojo/Week2/CalIt2.data.csv')
df
```

Se cambia el nombre de las columnas

```python
df = df.rename(columns={"0": "ID_Flujo", "1": "Fecha", "2": "Hora", "3": "Conteo"})
df.head()
```

Se seleccionan los datos del 24 de julio de 2005

```python
 Julio_24 = df.loc[df["Fecha"] == "07/24/05", :]
 Julio_24
```

Se seleccionan los datos con número de identificación de flujo = 7

```python
Flujo_7 = df.loc[df["ID_Flujo"] == 7, :]
Flujo_7
```

Se seleccionan los datos del 24 de julio de 2005 con número de identificación de flujo = 7

```python
Jul_24_ID_7 = df.loc[(df["Fecha"] == "07/24/05") & (df["ID_Flujo"] == 7), :]
Jul_24_ID_7
```

Se clasifican los valores de la columna "Conteo" y se seleccionan los 10 mayores

```python
Top10 = Jul_24_ID_7.sort_values(by="Conteo", ascending=False).head(10)
Top10
```

Se identifica el valor minimo de la columna "Conteo"

```python
Min = (Top10["Conteo"].min())
Min
```

Se substrae el valor minimo a todos los valores de la columna "Conteo" de Top10

```python
New_Top10 = Top10["Conteo"].apply(lambda x: x-Min)
New_Top10
```

Se obtiene el valor promedio de la columna "Conteo"

```python
New_Top10.mean()
```

Se seleccionan los datos de Agosto de 2005 con número de identificación de flujo = 7

```python
Ag_ID_7 = df.loc[(df["Fecha"] >= "08/01/05") & (df["Fecha"] <= "08/31/05") & (df["ID_Flujo"] == 7), :]
Ag_ID_7
```

Se agrupan los datos por fecha y se destaca el maximo conteo

```python
Ag_ID_7.groupby(["Fecha"])[["Conteo"]].max()
```
